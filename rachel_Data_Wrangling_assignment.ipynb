1. Read the abstract. What is this paper about?

The article is talking about “data tidying” and the benefits of “tidy datasets” as a consistent data structure which makes data easy to model and visualize.

2. Read the introduction. What is the "tidy data standard" intended to accomplish?

The "tidy data standard" is intended to develop a set “data analysis tools that work well together” so that when individuals are cleaning data, particularly messy real world data, they are able to perform data analysis without spending all their time cleaning the data.

3. Read the intro to section 2. What does this sentence mean: "Like families, tidy datasets are all alike but every messy dataset is messy in its own way." What does this sentence mean: "For a given dataset, it’s usually easy to figure out what are observations and what are variables, but it is surprisingly difficult to precisely define variables and observations in general."

The sentence “"Like families, tidy datasets are all alike but every messy dataset is messy in its own way" means that because “tidy datasets are all alike” they are a lot easier to do analytics on because packages and analytics are build under the assumption that data will be in this format, whereas because “every messy dataset is messy in its own way” it is difficult/maybe impossible to do analytics on messy data without cleaning it.

The sentence "For a given dataset, it’s usually easy to figure out what are observations and what are variables, but it is surprisingly difficult to precisely define variables and observations in general" highlights the importance of taking time to think about what are observations and variables in a specific data set in context, rather than relying on generalizations or assumptions about variables and observations generally.

4. Read Section 2.2. How does Wickham define values, variables, and observations?

An observation is a given unit in the data set whose attributes are being measured/described. A variable is measuring a common thing across each of the observations. A specific variable for a specific observation is a value. Thinking about an example of this really helped me understand these definitions so, for instance, Wickham explains that people could be the observations, and in this case height could be a variable, so the specific height of a specific person would be considered a value.

5. How is "Tidy Data" defined in section 2.3?

Tidy data is a particular way of matching rows, columns, and tables with observations, specifically in tidy data “each variable forms a column”,  “each observation forms a row” and, “each type of observational unit forms a table.” In other words, each table can only represent one observational unit, and within that table the variables must each be stored in a column.

6. Read the intro to Section 3 and Section 3.1. What are the 5 most common problems with messy datasets? Why are the data in Table 4 messy? What is "melting" a dataset?

Wickham explains that the 5 most common problems are “Column headers are values, not variable names”,  “Multiple variables are stored in one column”, “Variables are stored in both rows and columns”, “Multiple types of observational units are stored in the same table” and, “A single observational unit is stored in multiple tables”. The data in table 4 are messy because the column headers are values, in this case income ranges, not variable names. Melting data stacks the data and converts columns into rows

7. Why, specifically, is table 11 messy but table 12 tidy and "molten"?

Table 11 is messy because the column headers for each possible day in the month are values. Table 12a is molten because each day is melted into one column “date”. However, the element column represents multiple variables, which is resolved in table 12b making it tidy.

8. Read Section 6. What is the "chicken-and-egg" problem with focusing on tidy data? What does Wickham hope happens in the future with further work on the subject of data wrangling?

Because tools are developed based around standards for cleaning/tidying data, then those tools are only useful for data that are structured in a particular way. Because of this it is difficult to think about changing/improving this by  exploring “alternative formulations of tidiness” because it is unlikely that the existing tools that work with “tidy datasets” will work with alternative ways of tidying data.

Wickham hopes that others will build on their framework “to develop even better data storage strategies and better tools” rather than feeling limited in their data cleaning/tidying due to the “ the tidy data framework”

```python
df1=pd.read_csv('/content/wrangling/assignment/data/airbnb_hw.csv')
#This code shows the title of each column and information about the type of variable for each
print(df1.columns, '\n')
df1.head()
print(df1.info())
```

```python
# coerce a variable to numeric:
df4['imposed_sentence'] = pd.to_numeric(df4['imposed_sentence'], errors='coerce') # Coerce the variable to numeric

# Create a missing dummy:
df4['sentence_nan'] = df4['imposed_sentence'].isnull() # Equals 1 if missing, 0 if non-null

# see how many missing values there are coercion:
print('After coercion: \n', df4['imposed_sentence'].describe(),'\n') # Describe the numeric variable
df4['imposed_sentence'].hist(bins = 50) # Histogram of the variable values
print('Total Missings: \n', sum(df4['sentence_nan']),'\n') # Total missings
```

```python
# according to the code book if the sentence type is 0 that means "Probation/no incarceration"
#also if sentence type is 4 then that means "Other, all charges in cohort contact resulted in pending, dismissed, deferred, etc"
#for both of these the person ended up with no jail time so I used that to clean the 'imposed_sentence' variable
df4.loc[ df4['sentence_type'] == 0, 'imposed_sentence'] = 0
df4.loc[ df4['sentence_type'] == 4, 'imposed_sentence'] = 0
```
